## Drop useless columns

```python
df.drop('id', axis=1, inplace=True)
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['year'] = df['date'].apply(lambda date: date.year)
```

```python
df['month'] = df['date'].apply(lambda date: date.month)
```

```python
sns.boxplot(x='month', y='price', data=df)
```

```python
df.groupby('month').mean()['price'].plot()
```

```python
# drop engineered features
df.drop('date', axis=1, inplace=True)
```

```python
df['zipcode'].value_counts()
```

70 dummy variables are too many, so we will drop the ones that have a low correlation with the price

```python
df.drop('zipcode', axis=1, inplace=True)
```

```python
X = df.drop('price', axis=1).values
```

```python
y = df['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(
    Dense(
        # number of neurons same as number of features
        19,
        # rectified linear unit
        activation='relu'
    )
)
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))

# output layer
model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(
    x=X_train,
    y=y_train,
    # validation data is used to evaluate the loss function
    validation_data=(X_test, y_test),
    # batch size is the number of samples per gradient update.
    # the lower the batch size, the longer it takes to train.
    batch_size=128,
    epochs=400
)
```

```python
# dataframe that represents the loss function over the epochs for the training and test data
# this makes us able to understand if the model is overfitting or underfitting
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
# the loss function for the training and test data are very close, which means the model is not overfitting or underfitting
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score
```

```python
p = model.predict(X_test)
```

```python
mean_squared_error(y_test, p)
# it's a high value, but it's because we're squaring the price, which is a high value 
```

```python
# root mean squared error
np.sqrt(mean_squared_error(y_test, p))
```

```python
mean_absolute_error(y_test, p)
```

```python
# 
explained_variance_score(y_test, p)
```

```python
plt.scatter(y_test, p)
plt.plot(y_test, y_test, 'r')
# it seems like the model is doing a good job, but it's not predicting the higher prices very well
```
